Python also knows the operators & and | that are similar to 'and' and 'or'

Be careful, they are not the same! & and | are bitwise operators, while 'and' and 'or' are logical operators. Using bitwise operations for vector comparison with objects like lists might lead to unexpected results. We will encounter lists later in this tutorial. We recommend to use 'and' and 'or' by default.

```python
a, b = True, False
print(a & b)
print(a | b)
print((a | b) & (a | b))
print((a | b) | (b | b))
```

## Sets

-  A set is a list with no duplicate elements, so only unique elements. 
-  If an element already exists in the set it is not added. 
-  The set is unordered, so subsetting by index is useless.  
-  A set is created with the function set(). 
-  For sets there are useful operations such as intersection and difference.

```python
x = set(["a", "b", "c"])
type(x)
```

```python
x = {"a", "b", "c"}
type(x)
```

```python
# Create a set
x = set(["a", "b", "c"])

# Add an element to a set
x.add("d")
print(x)

# Add multiple elements to a set with the update() function
x.update(["4", "2"])
print(x)

# Remove an element from the set (remove the element by name, not by index)
x.discard("b")
print(x)
```

```python
dir(set)
```

```python
# A set consists of unique elements, so duplicates are removed automatically.
print(set("my name is Christian and Christian is my name".split()))
```

```python
# Create sets
a = set(["Jake", "John", "Eric"])
b = set(["John", "Jill"])

# The intersection of two sets:
print(a.intersection(b))

# The difference between two sets:
print(a.difference(b))
print(a-b)
```

# Advanced: Zipping & Unpacking

```python
countries_keys = list(countries_names.keys())
print(countries_keys)
countries_values = list(countries_names.values())
countries_values
```

```python
countries_tuples = dict(zip(countries_values, countries_keys))
countries_tuples
```

```python
myDict = dict()

for country_key, country_value in zip(countries_keys, countries_values):
  myDict[country_value] = country_key

myDict
```

```python
{country_value: country_key for country_key, country_value in zip(countries_keys, countries_values)}
```

```python
a, *b, c = 0, 1, 2, 3
print(a)
print(b)
print(c)
```

```python
countries_first, countries_second, *countries_other, countries_last = countries_values
print('countries_first:', countries_first)
print('countries_second:', countries_second)
print('countries_other:', countries_other)
print('countries_last:', countries_last)
```

```python
countries_first, countries_second, *countries_other, countries_last = countries_names.items()
print('countries_first:', countries_first)
print('countries_second:', countries_second)
print('countries_other:', countries_other)
print('countries_last:', countries_last)
```
